# Plotting with Python

This notebook introduces plotting with Python using Matplotlib.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## A Simple Plot
 
We begin with a simple sine plot using `np.sin`. First we use `np.linspace` to create a list of all our X points. In this case, an evenly spaced list from $0$ to $4\pi$ with 100 points. We then generate our Y points by calling `np.sin` on the X list. Finally, we can use `plt.plot(x, y)` to plot the results.


```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

## Multiple Lines

Now we'll plot multiple lines on a s single chart. In this case, 
- $y_1 = \sin{(x)}$
- $y_2 = \cos{(x)}$
- $y_3 = 2\sin{(x)}$ 
- $y_4 = 2\cos{(x)}$. 

We have the option of calling `plt.plot` once as `plt.plot(x, y1, x, y2, x, y3, x, y4)` or once for each plot (shown below). Note, we'll also specify the format
of the lines.

```python
x = np.linspace(0, 4*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = 2*np.sin(x)
y4 = 2*np.cos(x)

plt.plot(x, y1, '-b')   # Solid, lue line
plt.plot(x, y2, '--r')  # Red, dashed line
plt.plot(x, y3, '.g')   # Dotted, green line
plt.plot(x, y4, '-ok')  # Black solid line with circle markers
plt.show()
```
